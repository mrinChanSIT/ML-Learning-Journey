```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(X, y, theta, alpha, num_iters):
    """
    Perform gradient descent to learn theta.

    X : (m,n)
    y : (m, )
    theta : (n, )
    alpha : learning rate
    num_iters : number of iterations
   
    Returns : theta, cost_history
    """
    m = len(y)
    J_history = []

    for i in range(num_iters):
        
        predictions = X @ theta

        error  = predictions - y
        
        gradient = (1/m) * X.T @ error

        theta = theta - alpha * gradient

        cost = (1/2*m) * np.sum(error**2)

        J_history.append(cost)

        if i % 100 == 0:
            print(f"Iteration {i} : cost = {cost}")
        
    return theta, J_history
    
```

```python
np.random.seed(42)
X = np.c_[np.ones(100), np.random.randn(100)]
true_theta = np.array([1, 2])
y = X @ true_theta + np.random.randn(100) * 0.1
```

```python
theta = np.zeros(2)
alpha = 0.1
num_iters = 1000

```

```python
#Run Gradient Descent
theta, cost_history = gradient_descent(X, y, theta, alpha, num_iters)

print(f"\n Final theta: {theta}")
print(f"\n True Theta: {true_theta}")

print(f"Cost at convergence: {cost_history[-1]:.4f}")

# Visualize cost function
plt.plot(cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function')
plt.grid(True)
plt.show()



```

```python
# plot fitted line
plt.scatter(X[:,1], y, alpha=0.5)
x_line = np.linspace(X[:,1].min(), X[:, 1].max(), 100)
print(x_line.shape, x_line)
y_line = theta[0] + theta[1] * x_line
plt.plot(x_line, y_line, 'r-', label='Fitted Line')
plt.xlabel('X1')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
