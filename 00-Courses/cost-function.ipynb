```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(X, y, theta):
    """
    compute cost for linear regression

    X : (m,n) - m - examples, n features
    y : (m,) - m - target values

    theta  : (n, ) - features

    returns : cost (scalar)
    """

    m = len(X)

    predictions = X @ theta

    error = predictions - y
    print("error.shape", error.shape)
    # print("error.shape", error.shape)
    cost  = (1 / (2*m) ) * np.sum(error**2)

    return cost    
```

```python
# Test it
# Simple example: y = 2x + 1
np.random.seed(42)
X = np.c_[np.ones(100), np.random.randn(100)]
true_theta = np.array([1,2])
y = X @ true_theta + np.random.randn(100) * 0.1 

# Compute cost for different values of theta
# thetas = np.linspace(-1, 3, 100)

```

```python
theta_test = np.array([1, 2]) # Bias - 1, Slope - 2
print(f"Cost with theta=[1,2]: {cost_function(X, y, theta_test)}")
```

```python
# Visualize cost function for different slopes (keep bias=1)
slopes = np.linspace(0, 4, 100)
print("slopes.shape", slopes.shape)
costs = [cost_function(X, y, np.array([1, slope])) for slope in slopes]

plt.plot(slopes, costs)
plt.xlabel('Slope (theta_1)')
plt.ylabel('Cost')
plt.title('Cost Function vs Slope')
plt.axvline(x=2, color='r', linestyle='--', label='True slope')
plt.legend()
plt.grid(True)
plt.show()
```
