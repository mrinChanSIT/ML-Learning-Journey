```python
# Exercise 1: Vector operations
def vector_add(v1, v2):
    """Add two vectors"""
    # YOUR CODE
    #v_sum = v1 + v2
    v_sum = v1 + v2
    return v_sum
    pass
```

```python
def vector_scale(v, scalar):
    """Scale a vector"""
    # YOUR CODE
    #v_scl = scalar * v
    v_scl = scalar * v
    # print(type(v_scl))
    return v_scl
    pass
```

```python
# Exercise 2: Linear transformation
def apply_transformation(matrix, vector):
    """Apply linear transformation to a vector"""
    # YOUR CODE - do NOT use @ or np.dot
    # linear trabsformation is nothing but scaling a vector and adding it.
    # X = []
    # Y = []
    # for row in matrix:
    #     for i in range(len(row)):
    #         if  i == 0:
    #             X.append(row[i])
    #         else:
    #             Y.append(row[i])
                
    # I,J = np.array(X), np.array(Y)
    # r_90 = np.array([I,J])
    # tx_im = []
    # for i in range(len(vector)):
    #     tx_im.append(vector[i] * r_90[i])
    # return sum(tx_im)
    # pass
    
    # Extract columns (transformed basis vectors)
    basis_vectors = matrix.T  # Transpose to get columns as rows
    
    # Linear combination: sum of (scalar * basis_vector)
    result = sum(vector[i] * basis_vectors[i] for i in range(len(vector)))
    
    return result.tolist() if isinstance(result, np.ndarray) else result
```
