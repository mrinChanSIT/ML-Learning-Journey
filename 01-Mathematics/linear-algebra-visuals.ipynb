```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def visualize_transformation(matrix, title=""):
    """Visualize how a matrix transforms the unit square"""
    # Unit square vectors
    i_hat = np.array([1, 0])
    j_hat = np.array([0, 1])
    
    # Original square
    square = np.array([[0, 1, 1, 0, 0],
                       [0, 0, 1, 1, 0]])
    
    # Apply transformation
    transformed = matrix @ square
    
    # Plot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Original
    ax1.plot(square[0], square[1], 'b-', linewidth=2)
    ax1.arrow(0, 0, i_hat[0], i_hat[1], color='r', width=0.05)
    ax1.arrow(0, 0, j_hat[0], j_hat[1], color='g', width=0.05)
    ax1.set_xlim(-2, 2)
    ax1.set_ylim(-2, 2)
    ax1.grid(True)
    ax1.set_title("Original")
    ax1.set_aspect('equal')
    
    # Transformed
    new_i = matrix @ i_hat
    new_j = matrix @ j_hat
    ax2.plot(transformed[0], transformed[1], 'b-', linewidth=2)
    ax2.arrow(0, 0, new_i[0], new_i[1], color='r', width=0.05)
    ax2.arrow(0, 0, new_j[0], new_j[1], color='g', width=0.05)
    ax2.set_xlim(-2, 2)
    ax2.set_ylim(-2, 2)
    ax2.grid(True)
    ax2.set_title(f"After {title}")
    ax2.set_aspect('equal')
    
    plt.tight_layout()
    plt.show()

# Try different transformations
print("Rotation by 90 degrees:")
rotation_90 = np.array([[0, -1], [1, 0]])
visualize_transformation(rotation_90, "90° rotation")

print("\nShear transformation:")
shear = np.array([[1, 1], [0, 1]])
visualize_transformation(shear, "shear")

print("\nScaling:")
scale = np.array([[2, 0], [0, 2]])
visualize_transformation(scale, "2x scale")
```
